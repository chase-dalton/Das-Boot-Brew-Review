```python
# Import dependencies
import pandas as pd
```

```python
# read in the dataset
xml_df = pd.read_csv("beer_xmls.csv")
xml_df = xml_df.drop(columns = "Unnamed: 0")
xml_df.head()
```

```python
# get a list of all 42 column names
columns = xml_df.columns.tolist()
columns
```

```python
# Find out how many are 'award winning'Clean_data
# Can we predict if it will win an award based on any metric?
xml_df.value_counts("AWARD_WINNING")
```

```python
# Get info
xml_df.info()
```

```python
# Count null values
xml_df.isna().sum()
```

```python
# for AWARD_WINNING - change null to 0
xml_df["AWARD_WINNING"] = xml_df["AWARD_WINNING"].fillna(0)
```

```python
# Count null values
xml_df.isna().sum()
```

```python
# Check for and count duplicate rows
len(xml_df)-len(xml_df.drop_duplicates())
```

```python
# drop duplicate rows
xml_df = xml_df.drop_duplicates()
```

```python
# change "true" to 1 in the AWARD_WINNING column
xml_df["AWARD_WINNING"].replace("True", "1", inplace = True)
xml_df["AWARD_WINNING"] = xml_df.AWARD_WINNING.astype(float)
xml_df.head()
```

```python
# Count null values
xml_df.isna().sum()
```

```python
# Should we drop columns with lots of nulls?
xml_nonnull = xml_df.loc[:, xml_df.isnull().mean() < .5]
xml_nonnull.info()
```

```python
# Now drop rows with null
xml_nonnull = xml_nonnull.dropna()
xml_nonnull.info()
```
