```python
# import libraries
import pandas as pd
import os
```

```python
# loop through xml documents, import data into DF and merge into beers_xml_df
dir = r'.\bf\bf'

beers_xml_df = pd.DataFrame()

for file in os.listdir(dir):
    if file.endswith(".xml"):
        try:
            df = pd.read_xml(os.path.join(dir,file))
            beers_xml_df = pd.concat([beers_xml_df, df], axis=0, ignore_index=True)
        except:
            print(f'{file} had an Error.')
    else:
        continue
```
